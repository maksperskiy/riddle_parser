```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://heaclub.ru/zagadki-na-logiku-luchshaya-podborka-dlya-detej-i-vzroslyh-120-zagadok"
r = requests.get(url)
soup = BeautifulSoup(r.text, "html.parser")
```

```python
print(soup.prettify())
```

```python
article = soup.find('article')
article
```

```python
ols = article.find_all('ol')
ols
```

```python
lis = []
[lis.extend(li.find_all('li')) for li in ols]
lis
```

```python
riddles = {li.text.replace(li.find('strong').text, '').strip(): li.find('strong').text.replace('(Ответ — ', '').replace('.)', '').strip() for li in lis}
```

```python
riddles
```

```python
with open('riddles.txt', 'w') as f:
    for key, value in riddles.items():
        f.write(key + "; " + value + "\n")
```
